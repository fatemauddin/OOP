```python
import csv
```

```python
def read_csv_as_list(filename):
    data = []
    
    with open(filename, mode='r', newline='', encoding='utf-8') as file:
        reader = csv.reader(file)
        
        for row in reader:
            data.append(row)
    
    return data
```

```python
rows = read_csv_as_list("salarysheet.csv")
rows
```

```python
max_row = max(rows, key=lambda x: x[1])
print(max_row)
```

```python
min = min(rows, key=lambda x: x[0])
print(min)
```

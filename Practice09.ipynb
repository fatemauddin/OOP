```python
#BankAccount class
class BankAccount:
    def __init__(self, account_number, owner_name, balance=0.0):
        self._account_number = account_number
        self._owner_name = owner_name
        self._balance = balance

    @property
    def account_number(self):
        return self._account_number

    @property
    def owner_name(self):
        return self._owner_name

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit amount must be positive")
        self._balance += amount

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Withdrawal amount must be positive")
        if amount > self._balance:
            raise ValueError("Insufficient balance")
        self._balance -= amount

    def get_balance(self):
        return self._balance
```

```python
#Bank Class
class Bank:
    def __init__(self):
        self._accounts = []

    def add_account(self, account):
        self._accounts.append(account)

    def total_bank_balance(self):
        return sum(account.get_balance() for account in self._accounts)

    def show_all_accounts(self):
        print("\nBank Accounts:")
        for acc in self._accounts:
            print(f"{acc.owner_name} | Acc No: {acc.account_number} | Balance: {acc.get_balance()}")
```

```python
#Objects
bank = Bank()

# Create bank accounts
acc1 = BankAccount("A001", "John Doe", 1000)
acc2 = BankAccount("A002", "Jane Smith", 2000)
acc3 = BankAccount("A003", "Alice Brown", 1500)

# Add accounts to bank
bank.add_account(acc1)
bank.add_account(acc2)
bank.add_account(acc3)

# Perform transactions
acc1.deposit(500)
acc2.withdraw(300)
acc3.deposit(700)
acc1.withdraw(200)

# Show all accounts
bank.show_all_accounts()

# Calculate total balance owned by bank
print("\nTotal balance owned by bank:", bank.total_bank_balance())
```
